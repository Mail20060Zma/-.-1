Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def differ(list1,list2):
  temp=[]

  for i in range(len(list1)):
    if list1[i] not in list2:
      temp.append(list1[i])

  for i in range(len(list2)):
    if list2[i] not in list1:
      temp.append(list2[i])
  return temp

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print(differ(a,b))
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
pets_dict={}

def new_pet():
  name = input('Введите имя питомца: ').lower().rstrip()
  types = input('Введите тип питомца: ')
  age = int(input('Введите возраст питомца: '))
  pets_dict[name] = {'type': types, 'age': age, 'status' : {'hunger': 50, 'joy': 50, 'energy': 50}}



def play_pet():
  for i in pets_dict:
    print(i)

  name = input('Выберите питомца с которым хотите взаимодействовать ').lower().rstrip()
  if name in pets_dict:
    print('''
    1 - покормить питомца
    2 - поиграть с питомцем
    3 - уложить спать ''')
    action = input('  Выберите действие: ')


    if action == '1':
      if pets_dict[name]["status"]["hunger"] < 100:
        print(f'Сытость питомца выросла на 10 и теперь она ровна {pets_dict[name]["status"]["hunger"] + 10}')
        pets_dict[name]["status"]["hunger"] = pets_dict[name]["status"]["hunger"] + 10
      else:
        print('Сытость питомца на максимуме!')


    if action == '2':
      if pets_dict[name]["status"]["hunger"] >= 5 and pets_dict[name]["status"]["energy"]>=5:
        print('Вы поиграли с питомцем и его радость выросла на 10, а голод и энергией уменшились на 5!')

        if pets_dict[name]["status"]["joy"] >= 90 :
          pets_dict[name]["status"]["joy"] = 100
          print('Радость вашего питомца максимальна!')
        else:
          pets_dict[name]["status"]["joy"] = pets_dict[name]["status"]["joy"] + 10

        pets_dict[name]["status"]["energy"] = pets_dict[name]["status"]["energy"] - 5
        pets_dict[name]["status"]["hunger"] = pets_dict[name]["status"]["hunger"] - 5
      else:
        print('Вашему питомцу не хвататет сытости или энергии!')


    if action == '3':
      if pets_dict[name]["status"]["hunger"] >= 10 and pets_dict[name]["status"]["joy"] >= 5:
        print('Ваш питомец поспал и востоновил 15 энергии и потерял 10 голода и 5 радости!')

        if pets_dict[name]["status"]["energy"] >= 85:
          pets_dict[name]["status"]["energy"] = 100
          print('энергия вашего питомца максимальна!')
        else:
          pets_dict[name]["status"]["energy"] = pets_dict[name]["status"]["energy"] + 15

        pets_dict[name]["status"]["hunger"] = pets_dict[name]["status"]["hunger"] - 10
        pets_dict[name]["status"]["joy"] = pets_dict[name]["status"]["joy"] - 5
      else:
        print('Вашему питомцу не хватает сытости или радости!')



def status_pet():
  for i in pets_dict:
    print(i)

  name = input('Выберите питомца с которым хотите взаимодействовать ').lower().rstrip()
  if name in pets_dict:
    print(f'''  У {name}
    Возрост - {pets_dict[name]["age"]}
    Тип - {pets_dict[name]["type"]}
    Сытость - {pets_dict[name]["status"]["hunger"]}
    Радость - {pets_dict[name]["status"]["joy"]}
    Энергия - {pets_dict[name]["status"]["energy"]}''')



new_pet()

while True:
  print('''
  1 - Завести нового питомца
  2 - Взаимодействовать с питомцем
  3 - Статус питомца
  4 - Выход ''')
  action = input('  Выберите действие: ')

  if action == '1':
    new_pet()

  elif action == '2':
    play_pet()

  elif action == '3':
    status_pet()

  elif action == '4':
    break
````

````python
from random import randint


playr_dict = {'money': 50000,
              'stats': {'HP': 100, 'strang': 100, 'endurance': 100, 'defens':0, 'damage': 5 , 'damage_rang':0},
              'inventory':{}
              }


data_products = {'armor':{'iron_helmet':{'price': 10000, 'defens': 10, 'description':'Защищает голову!'},
                          'iron_chestplate':{'price': 30000, 'defens': 20,'description':'Защищает тело!'},
                          'shield':{'price': 10000, 'defens': 5,'description':'Защищает все!'},
                          'iron_legins':{'price': 15000, 'defens': 10,'description':'Защищает ноги!'}},

                'weapon':{'iron_sword':{'price': 10000, 'damage': 15,'description':'Атакует в близи можете получть ответный удар!'},
                          'bow':{'price': 15000, 'damage': 10,'description':'Атакует в дали на безапасном расстояние!'}},

                'pets':{'dog_pet':{'price': 10000, 'damage': 5,'description':'Атакует каждых ход'},
                        'crocodile_pet':{'price': 30000, 'damage': 20,'description':'Атакует каждых ход'},
                        'dragon_pet':{'price': 75000, 'defens': 30,'description':'Атакует каждых ход'}}
            }

def new_game():
  name = input ('  Выберите имя персонажа: ')
  playr_dict['name'] = name
  products = data_products

  score = 100
  print(f'''  У вас всего 100 очков нывыков и {playr_dict['money']} монет!''')

  while True :
    print(f'''  У вас {score} очков!
    1 - Вложить очки в силу
    2 - Вложить очки в HP
    3 - Вложить очки в выносливоть
    4 - Вложить очки в деньги ''')

    action = input('Выберите действие: ')
    if action == '1':
      temp = int(input('Выберите количесвто очков которые хотите вложить: '))
      if temp != 0 and score >= temp :
        score = score - temp
        playr_dict['stats']['strang'] = playr_dict['stats']['strang'] + temp

    elif action == '2':
      temp = int(input('Выберите количесвто очков которые хотите вложить: '))
      if temp != 0 and score >= temp:
        score = score - temp
        playr_dict['stats']['HP'] = playr_dict['stats']['HP'] + temp

    elif action == '3':
      temp = int(input('Выберите количесвто очков которые хотите вложить: '))
      if temp != 0 and score >= temp:
        score = score - temp
        playr_dict['stats']['endurance'] = playr_dict['stats']['endurance'] + temp

    elif action == '4':
      temp = int(input('Выберите количесвто очков которые хотите вложить: '))
      if temp != 0 and score >= temp:
        score = score - temp
        playr_dict['money'] = playr_dict['money'] + temp*1000
    if score == 0:
      break


  while True:
    print(f'0 - Назад')
    for i in range(len(list(products.keys()))):
      print(f'{i+1} - {list(products.keys())[i]}')

    action = int(input('Выберите действие: '))
    if action == 0: break
    data = products[list(products.keys())[action-1]]

    while True:
      print(f'0 - Назад')

      for i in range(len(list(data.keys()))):
        print(f'''{i+1} - {list(data.keys())[i]}
        price - {data[list(data.keys())[i]]["price"]}
        {list(data[list(data.keys())[i]].keys())[1]} - {data[list(data.keys())[i]][list(data[list(data.keys())[i]].keys())[1]]}
        description - {data[list(data.keys())[i]]["description"]}''')

      action = int(input('Выберите предмет который хотите купить: '))
      if action == 0: break
      elif action != 0 and playr_dict['money'] >= data[list(data.keys())[action-1]]['price'] :
        playr_dict['money'] = playr_dict['money'] - data[list(data.keys())[action-1]]['price']
        print(f'''
        {list(data.keys())[action-1]} был куплен и добален в инвентарь!
        У вас осталось {playr_dict['money']} денег''')
        playr_dict['inventory'][list(data.keys())[action-1]] = (data[list(data.keys())[action-1]])
      else:
        if playr_dict['money'] >= data[list(data.keys())[action-1]]['price'] : print('Вам не хватает денег!')

  data=playr_dict['inventory']
  for i in range(len(list(data.keys()))):
    print(f'''{i+1} - {list(data.keys())[i]}
          {list(data[list(data.keys())[i]].keys())[1]} - {data[list(data.keys())[i]][list(data[list(data.keys())[i]].keys())[1]]}
          description - {data[list(data.keys())[i]]["description"]}''')

  print('''  Вы просыпаетесь в товерни и берети задание на убийство дракона и идешь на его убийство!
  По пути на вас наподает одинокого Гоблина!''')
  goblin = {'HP': 50, 'damage':15}

  while True:
    print(f'''    У вас {playr_dict["stats"]["HP"]} HP
    У гоблина {goblin["HP"] } HP
    1 -  Атаковать
    ''')
    action = input('')
    if action == '1':
      goblin['HP'] -= playr_dict['stats']['damage']*(playr_dict['stats']['strang']/100)
      print(f'Вы нанесли удар и  у гоблина осталось {goblin["HP"]} HP')
      if goblin['HP'] <= 0:
        print('Вы победили гоблина!')
        break
      else:
        playr_dict['stats']['HP'] -= goblin['damage']
        print(f'Гоблин атакует вас и  у вас осталось {playr_dict["stats"]["HP"]} HP')
        if playr_dict['stats']['HP'] <= 0:
          print('Вы проиграли гоблину!')
          break
    elif action == '2':
      print('Вы сбежали!')
      break

  print('Вы продолжаете путь к дракону.')

  dragon = {
    'HP': 200,
    'damage': 30
  }

  while True:
    print(f'''    У вас {playr_dict["stats"]["HP"]} HP
    У дракона {dragon["HP"]} HP
    1 -  Атаковать
    2 - Бежать
    ''')
    action = input('')
    if action == '1':
      dragon['HP'] -= playr_dict['stats']['damage']*(playr_dict['stats']['strang']/100)
      print(f'Вы нанесли удар и  у дракона осталось {dragon["HP"]} HP')
      if dragon['HP'] <= 0:
        print('Вы победили дракона!')
        break
      else:
        playr_dict['stats']['HP'] -= dragon['damage']
        print(f'Дракон атакует вас и  у вас осталось {playr_dict["stats"]["HP"]} HP')
        if playr_dict['stats']['HP'] <= 0:
          print('Вы проиграли дракону!')
          break
    elif action == '2':
      print('Вы сбежали!')
      break



# Как говриться планы были большие, а получилось говно я не успел и третье мне лень

new_game()
````
